```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.simplefilter('ignore')

df = pd.read_excel('/Users/seanyu_mac/Downloads/주문내역_2024-03-17~2025-03-17 (3).xlsx')

con1 = df['주문 상태'] == '결제완료'
con2 = df['주문 상태'] == '배송준비'
con3 = df['주문 상태'] == '배송중'
con4 = df['주문 상태'] == '배송완료'
con5 = df['상품 번호'] == 51174

df = df[con5]
df = df[con1 | con2 | con3 | con4]


df.rename(columns={'결제 완료일': '결제 시점'}, inplace=True)

now = datetime.now().strftime('%Y-%m-%d %H:%M분')

print(f'{now} 현재 시간 총 매출은 {df['정산가'].sum():,}원')
print(f'{now} 현재 시간 총 판매건수는 {df.index.size}건')
result_df = pd.DataFrame(df.groupby(pd.Grouper(key='결제 시점', freq='30T'))['정산가'].sum())
result_df['정산가'] = result_df['정산가'].apply(lambda x: f"{x:,} 원")
result_df.rename(columns={'정산가': '매출'})
```
