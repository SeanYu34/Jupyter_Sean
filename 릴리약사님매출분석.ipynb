```python
df.sort_values(by='order_amount', ascending=False)[:10]
```

```python
# ✅ 5천만 원 단위로 구간을 설정
bins = list(range(0, df['order_amount'].max() + 50000000, 50000000))

# ✅ 구간에 대한 라벨(이름) 만들기
labels = [f"{bins[i]}~{bins[i+1]-1}" for i in range(len(bins)-1)]

# ✅ 매출 금액을 해당 구간에 맞게 분류
df['sales_range'] = pd.cut(df['order_amount'], bins=bins, labels=labels, right=False)

# ✅ 각 구간별 판매 건수 계산
sales_count = df['sales_range'].value_counts().sort_index()

df['sales_range'].value_counts()
```
