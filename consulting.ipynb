```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

df = pd.read_csv('/Users/seanyu_mac/Downloads/public_consulting_export_2025-01-06_115107.csv')
pharmacist_df = pd.read_csv('/Users/seanyu_mac/Downloads/public_admin_export_2025-01-06_143622.csv')
df.columns
```

```python
df['reply_yn'].unique()
```

```python
df = df[['id', 'user_id', 'referer', 'reg_dt', 'del_yn', 'answer_dt', 'public_yn', 'recommend_product_ids', 'answer_user_id', 'reply_yn', 'answer_pharmacy_id']]
```

```python
df.head(3)
```

```python
pharmacist_df.rename(columns={'id' : 'answer_pharmacy_id'}, inplace=True)
pharmacist_df
```

```python
df['reg_dt'] = df['reg_dt'].str[:10]
df['reg_dt'] = pd.to_datetime(df['reg_dt'])

df['answer_dt'] = df['answer_dt'].str[:10]
df['answer_dt'] = pd.to_datetime(df['answer_dt'])
```

```python
# 월별 상담건수 (2024년)
# 일자별 평균 상담건수 (2024년)
# 월별 평균 답변 소요시간
# 답변 공개 비율
# 추천 제품 작성 비율
# 경로별 상단 인입 비율
```

```python
# 2024년에 등록된 상담만 따로 추출
con1 = df['reg_dt'] >= datetime(2024, 1, 1)
con2 = df['reg_dt'] <= datetime(2024, 12, 31)
df = df[con1 & con2]
df.drop(df[df['answer_dt'].isna()].index, inplace=True)
```

```python
df = pd.merge(df, pharmacist_df[['answer_pharmacy_id', 'name']], how='inner', on='answer_pharmacy_id')
df.head(3)
```

```python
# 월별 상담 건수 / 일자별 평균 상담 건수
df['reg_month'] = df['reg_dt'].dt.month
df1 = df.groupby('reg_month').size().reset_index(name='count')
df1['avg_daily_count'] = np.round(df1['count'] / 30, 0)
df1.rename(columns={'count' : 'avg_monthly_count'}, inplace=True)
df1
```

```python
# 월별 평균 답변 소요시간
df['taken_time'] = df['answer_dt'] - df['reg_dt']
df1['avg_taken_time'] = df.groupby('reg_month')['taken_time'].mean().tolist()
df1.to_excel('/Users/seanyu_mac/Downloads/2024_consult_overview.xlsx')
```

```python
# 상담 공개 비율
df['public_yn'].value_counts()[1] / df['public_yn'].value_counts()[0] * 100
```

```python
# 추천 제품 작성 비율
df[df['recommend_product_ids'].isna()].size / df.size * 100
```
