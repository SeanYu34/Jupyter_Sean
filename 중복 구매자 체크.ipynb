```python
df = pd.read_excel('/Users/seanyu_mac/Downloads/주문내역_2024-05-13~2025-05-13.xlsx')
```

```python
df['구매자 연락처'] = df['구매자 연락처'].astype(str)
df = df[df['상품명'].str.contains('공동')]
```

```python
df = df[df['주문 상태'].isin(['결제완료', '배송준비', '배송중', '배송완료', '구매확정'])]
```

```python
df = df.drop(df[df['구매자 연락처'] == 'nan'].index)
```
