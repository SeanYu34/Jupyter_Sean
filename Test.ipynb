```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
import matplotlib.pyplot as plt
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('/Users/seanyu_mac/Downloads/누적 매출_Master - raw.csv')
fs_df = pd.read_excel('/Users/seanyu_mac/Downloads/주문내역_2024-04-04~2025-04-04.xlsx')
```

```python
df.columns = df.loc[2]
df = df.drop([0,1,2]).reset_index()
df = df[df['채널명 (쇼핑몰명)'] == '친한 스토어']
```

```python
df = df[df['채널상품명'] == '0']
```

```python
fs_df = fs_df[fs_df['주문 상태'].isin(['결제완료', '배송중', '배송완료', '배송준비', '구매확정'])]
```

```python
df['실결제금액']
```

```python
fs_df['정산가'] = fs_df['정산가'].apply(lambda x: f'{x:,}')
fs_df['실결제금액'] = fs_df['정산가'].astype(str)
fs_df['실결제금액']
```

```python
df['주문번호'] = df['주문번호'].apply(lambda x: x[:5])
df['주문번호'] = df['주문번호'].astype(str)
df['주문번호']
```

```python
fs_df['주문 번호'] = fs_df['주문 번호'].apply(lambda x: x[:5])
fs_df['주문번호'] = fs_df['주문 번호'].astype(str)
fs_df['주문번호']
```

```python
df['주문일'] = pd.to_datetime(df['주문일'])
df['주문일'] = df['주문일'].astype(str)
df['주문일']
```

```python
fs_df['결제 완료일'] = fs_df['결제 완료일'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
fs_df['주문일'] = fs_df['결제 완료일']
fs_df['주문일']
```

```python
fs_df
```

```python
fs_df_dedup = fs_df.drop_duplicates(subset=['주문번호', '주문일'])

merge_df = pd.merge(df[['실결제금액', '주문번호', '주문일']], fs_df_dedup[['실결제금액', '주문번호', '주문일', '상품명', '옵션명', '상품 번호']], on=['주문번호', '주문일'], how='left')
```

```python
merge_df
```

```python
merge_df = merge_df.sort_values(by='주문일', ascending=True)
merge_df.to_excel('/Users/seanyu_mac/Downloads/주문내역345432543254-04.xlsx')
```
