```python
import pandas as pd

df = pd.read_csv('/Users/seanyu_mac/Downloads/[친한구매]약사별 판매링크 관리_친한스토어_250305 - 전체 (1).csv')
df
```

```python
product_nums = []

for i in df['링크']:
    product_num = i.split('/')[4].strip()
    product_nums.append(product_num)

df['product_num'] = product_nums
```

```python
for i in df['상품명'].unique():
    con = df['상품명'] == i
    price_list = df.loc[con, '정상가'].tolist() 
    for j in price_list:
        if j != price_list[0]:
            print(f'{i} 상품명 제품 중 {j} 원은 잘못된 정보입니다.')
```
