```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.simplefilter('ignore')

# 모든 행과 모든 컬럼을 표시하도록 설정
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
ob_df = pd.read_excel('/Users/mac/Downloads/품목_현황-250530.xlsx')
```

```python
fs_df = pd.read_csv('/Users/mac/Downloads/store_t_product_option_export_2025-05-30_085914.csv')
```

```python
pd_df = pd.read_csv('/Users/mac/Downloads/store_t_product_export_2025-05-30_092603.csv')
```

```python
ob_df.drop(ob_df[ob_df['업체품목코드'].isna()].index, inplace=True)
ob_df['업체품목코드'] = ob_df['업체품목코드'].astype(int)
```

```python
fs_df.drop(fs_df[fs_df['fulfillment_item_code'].isna()].index, inplace=True)
```

```python
fs_df['fulfillment_item_code'] = fs_df['fulfillment_item_code'].astype(int)
fs_df['fulfillment_item_code'] = fs_df['fulfillment_item_code'].astype(str)
del_list = []

for i in fs_df['fulfillment_item_code']:
    if len(i) < 12:
        del_list.append(i)
    elif len(i) > 12:
        del_list.append(i)
```

```python
fs_df.drop(fs_df[fs_df['fulfillment_item_code'].isin(del_list)].index, inplace=True)
fs_df['fulfillment_item_code'] = fs_df['fulfillment_item_code'].astype(int)
```

```python
fs_df.drop(fs_df[fs_df['status_yn'] == 'N'].index, inplace=True)
```

```python
fs_df_dict = fs_df.drop_duplicates(subset=['fulfillment_item_code'], keep='first')
ff_df = fs_df_dict.set_index('fulfillment_item_code')[['option_name', 'product_idx']]
ff_df['fulfillment_item_code'] = ff_df.index
```

```python
df = pd.merge(ff_df, pd_df[['product_idx', 'product_name']], on='product_idx', how='left')
```

```python
df = df[['fulfillment_item_code', 'product_idx', 'product_name']]
df.info()
```

```python
df = pd.merge(df, ob_df[['업체품목코드', '품목명']], left_on='fulfillment_item_code', right_on='업체품목코드', how='left')
```
