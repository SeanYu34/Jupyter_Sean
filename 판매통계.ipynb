```python
import numpy as np
from datetime import datetime, timedelta

dates = []
for i in df['order_ymd']:
    i = str(i)
    year = i[:4]
    month = i[4:6]
    day = i[6:]
    date = year + '-' + month + '-' + day
    dates.append(date)

df['order_ymd'] = dates
df['order_ymd'] = pd.to_datetime(df['order_ymd'], format='mixed')
```
