```python
import pandas as pd

excel_df = pd.read_excel('/Users/seanyu_mac/Downloads/20250110 1000 슈퍼맘약사 알래스카츄 주문내역.xlsx')
pb_df = pd.read_csv('/Users/seanyu_mac/Downloads/public_admin_group_buy_excel_order_export_2025-01-10_161638.csv')
con1 = excel_df['옵션명'] == '수량 : 알래스카 츄어블: 2박스 (4개월)'
con2 = excel_df['옵션명'] == '수량 : 알래스카 츄어블: 4박스 (8개월)'
con3 = excel_df['옵션명'] == '수량 : 알래스카 츄어블: 6박스 (12개월)'
excel_df['price'] = 0
excel_df['sales'] = 0
excel_df.loc[con1, 'price'] = 84000
excel_df.loc[con2, 'price'] = 159000
excel_df.loc[con3, 'price'] = 229000
excel_df['sales'] = excel_df['price'] * excel_df['구매수량']
```

```python
from datetime import datetime, timedelta

excel_df['주문일'] = pd.to_datetime(excel_df['주문일'])
excel_df
```

```python
df_30 = pd.DataFrame(excel_df.groupby(pd.Grouper(key='주문일', freq='30T'))['sales'].sum().reset_index())
df_60 = pd.DataFrame(excel_df.groupby(pd.Grouper(key='주문일', freq='1H'))['sales'].sum().reset_index())
df_180 = pd.DataFrame(excel_df.groupby(pd.Grouper(key='주문일', freq='3H'))['sales'].sum().reset_index())
df_360 = pd.DataFrame(excel_df.groupby(pd.Grouper(key='주문일', freq='6H'))['sales'].sum().reset_index())
df_day = pd.DataFrame(excel_df.groupby(pd.Grouper(key='주문일', freq='1D'))['sales'].sum().reset_index())
df_week = pd.DataFrame(excel_df.groupby(pd.Grouper(key='주문일', freq='W'))['sales'].sum().reset_index())
df_month = pd.DataFrame(excel_df.groupby(pd.Grouper(key='주문일', freq='M'))['sales'].sum().reset_index())
```

```python
with pd.ExcelWriter("/Users/seanyu_mac/Downloads/슈퍼맘약사시간별데이터.xlsx") as writer:
    df_30.to_excel(writer, sheet_name="df_30", index=False)
    df_60.to_excel(writer, sheet_name="df_60", index=False)
    df_180.to_excel(writer, sheet_name="df_180", index=False)
    df_360.to_excel(writer, sheet_name="df_360", index=False)
    df_day.to_excel(writer, sheet_name="df_day", index=False)
    df_week.to_excel(writer, sheet_name="df_week", index=False)
    df_month.to_excel(writer, sheet_name="df_month", index=False)
    
    
```
