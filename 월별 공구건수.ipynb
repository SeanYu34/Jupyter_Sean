```python
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.simplefilter('ignore')

df = pd.read_csv('/Users/seanyu_mac/Downloads/[공구] 일정 신청 및 관리 시트 - 공구전체일정 (3).csv')
df.columns = df.loc[1]
df.drop([0,1], inplace=True)
df.reset_index(drop=True, inplace=True)
del df['운영시트반영']

dates = []

for i in df['시작일']: 
    split = i.split('/')
    if len(split) >= 3:
        year = '20' + i.split('/')[0].strip()
        month = i.split('/')[1].strip()
        day = i.split('/')[2].strip()
        day = day.split('(')[0]
        date = year + '-' + month + '-' + day
    dates.append(date)

df['start_day'] = dates
df['start_day'] = pd.to_datetime(df['start_day'])

con1 = df['공급사'] == 'YDY뉴트리션'
con2 = df['공급사'] == '파이토알엑스'
con3 = df['공급사'] == '한독'
ydy_df = df[con1 | con2 | con3]
```

```python
con1 = df['start_day'] >= datetime(2025, 1, 1)
con2 = df['start_day'] <= datetime(2025, 12, 31)

df = df[con1 & con2]
ydy_df = ydy_df[con1 & con2]
```
