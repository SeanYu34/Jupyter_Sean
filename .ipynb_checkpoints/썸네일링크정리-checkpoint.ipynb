```python
df4 = df4[['상품명', '주문 상태', '결제 완료일', '정산가']]
df4['브랜드'] = '없음'
col = df4.columns.tolist()
col_name = ['상품명', '주문상태', '주문일', '정산기준금액', '브랜드']
cols = dict(zip(col, col_name))
df4.rename(columns=cols, inplace=True)
```

```python
df = pd.concat([kakao_df, df4])
df['주문상태'].unique()[[0, 6]].tolist()

con1 = df['주문상태'] == df['주문상태'].unique()[0]
con2 = df['주문상태'] == df['주문상태'].unique()[6]

df = df.loc[con1 | con2]
df.rename(columns={'브랜드' : '약사님'}, inplace=True)
df.reset_index(drop=True, inplace=True)
del_list = df[df['상품명'] == '임직원용 PB 상품 구매 페이지'].index.tolist()
```

```python
df.drop(del_list, inplace=True)
```

```python
con1 = df['약사님'].isna()
con2 = df['약사님'] == '없음'
df.loc[con1 | con2, '약사님'] = '없음'
df[con1 | con2]
```

```python
df['주문일'] = pd.to_datetime(df['주문일'])
df['주문일'] = df['주문일'].dt.date
df['약사님'].unique()
```

```python
con = df['상품명'] == '[공동구매] 우리약사 X 센스밸런스 헤모츄 / 2box, 4box, 8box'
df.drop([4467, 4468], inplace=True)
```

```python
df.loc[df['약사님'] == '없음']
```

```python
product_names = df.loc[df['약사님'] == '없음', '상품명'].tolist()
yaksa = []

for name in product_names:
    split = name.split('/')
    split2 = split[1].split(']')
    yaksa.append(split2[0].strip())

len(yaksa)
```

```python
df.loc[df['약사님'] == '없음', '약사님'] = yaksa
```

```python
df.loc[df['약사님'] == '친한', '약사님'] = '친한약사'
```

```python
df.rename(columns={'정산기준금액' : '판매액'}, inplace=True)
```

```python
sales = pd.DataFrame(df.groupby('약사님')['판매액'].sum().reset_index())
sales.sort_values(by='판매액', ascending=False, inplace=True)
sales.to_excel('/Users/seanyu_mac/Downloads/2024년_약사님별_상담매출.xlsx')
```
