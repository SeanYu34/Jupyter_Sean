```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.simplefilter('ignore')

df = pd.read_excel('/Users/seanyu_mac/Downloads/주문내역_2024-03-18~2025-03-18 (1).xlsx')
status = ['결제완료', '배송준비', '배송중', '배송완료', '구매확정']
df = df[df['주문 상태'].isin(status)]
df = df[df['상품명'].isin(df.loc[df['상품명'].str.contains('공동'), '상품명'].unique())]
result_df = df.groupby('상품명', as_index=False)['정산가'].sum()
result_df['정산가'] = result_df['정산가'].apply(lambda x: f'{x:,}원')
result_df
```

```python
result_df = df.groupby(['상품명', '옵션명'], as_index=False)['주문 번호'].count()
result_df.sort_values(by='주문 번호', ascending=False)
```
