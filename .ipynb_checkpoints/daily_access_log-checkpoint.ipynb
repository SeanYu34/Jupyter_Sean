```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

df = pd.read_csv('/Users/seanyu_mac/Downloads/public_daily_access_log_export_2024-11-27_130829.csv')
df.head(3)
```

```python
df['access_date'] = pd.to_datetime(df['access_date'], format='mixed')
df['created_at'] = pd.to_datetime(df['created_at'], format='mixed')
df['gap'] = df['created_at'] - df['access_date']
df['gap_seconds'] = df['gap'].dt.total_seconds()
df['gap_minutes'] = df['gap'].dt.total_seconds() / 60
df['gap_hours'] = df['gap'].dt.total_seconds() / 3600
df
```

```python
data1 = []
data2 = []

for i in range(df.index.size):
    # Get sorted, non-null values for the row
    data = df.loc[i].sort_values().dropna().tolist()
    
    if data:  # Ensure the row has valid data
        min_val = data[0]   # First (minimum) value
        max_val = data[-1]  # Last (maximum) value
        
        data1.append(min_val)
        data2.append(max_val)
    else:
        # Handle cases where the row is entirely NaN
        data1.append(None)
        data2.append(None)

len(data2)
```

```python
result = pd.DataFrame()
result['first'] = data1
result['last'] = data2
result['gap'] = result['last'] - result['first']
con1 = result['gap'] != timedelta(days=0)
con2 = result['last'] >= datetime(2024, 10, 26)
result[con1 & con2]
```

```python
temp = result[con1]
one_week_retention = temp[temp['last'] >= datetime(2024, 11, 19)].index.size
two_week_retention = temp[temp['last'] >= datetime(2024, 11, 12)].index.size
three_week_retention = temp[temp['last'] >= datetime(2024, 11, 5)].index.size

print(f'10월 7일 ~ 11월 26일까지\n1주 리텐션 {one_week_retention/temp.index.size}%\n2주 리텐션 {two_week_retention/temp.index.size}%\n3주 리텐션 {three_week_retention/temp.index.size}%\n')
```
