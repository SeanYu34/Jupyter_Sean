```python
import pandas as pd
import numpy as np

df = pd.read_excel('/Users/seanyu_mac/Downloads/누적판매금액 (25).xlsx')
df['order_amount'].mean() / 7
```

```python
df = df.sort_values(by='order_amount', ascending=False)
pivot_df = pd.pivot_table(df, index='sale_start_date', columns='group_buy_title', values='order_amount', aggfunc='sum', fill_value=0, margins=True, margins_name='총합')
pivot_df
```

```python
pd.DataFrame(df.groupby(['group_buy_title', 'sale_start_date'])['order_amount'].sum()).sort_values(by='sale_start_date')
```

```python
df['name'] = df['group_buy_title'].apply(lambda x: x.split(')')[0].split('(')[1].strip())
pd.DataFrame(df.groupby(['group_buy_title', 'name', 'sale_start_date'])['order_amount'].sum()).sort_values(by='order_amount', ascending=False)
```
