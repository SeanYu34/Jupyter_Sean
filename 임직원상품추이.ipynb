```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

df = pd.read_excel('/Users/seanyu_mac/Downloads/주문내역_2024-01-16~2025-01-16.xlsx')
df
```

```python
df = df[df['상품명'].str.contains('임직')]
df
```

```python
con1 = df['주문 상태'] == df['주문 상태'].unique()[2]
con2 = df['주문 상태'] == df['주문 상태'].unique()[3]
df = df[con1 | con2]
```

```python
df
```

```python
df['결제 완료일'] = pd.to_datetime(df['결제 완료일'])
result_df = pd.DataFrame(df.groupby(pd.Grouper(key='결제 완료일', freq='30T'))['정산가'].sum())
result_df.to_excel('/Users/seanyu_mac/Downloads/주문내역_2024-01-16~2025-01-161.xlsx')
```
