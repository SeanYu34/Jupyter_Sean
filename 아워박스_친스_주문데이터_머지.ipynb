```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.simplefilter('ignore')

ob_df = pd.read_excel('/Users/seanyu_mac/Downloads/아워박스 202502-03 친한스토어 raw.xlsx')
fs_df = pd.read_excel('/Users/seanyu_mac/Downloads/주문내역_2024-03-18~2025-03-18 (2).xlsx')
```

```python
ob_df['주문일시'] = pd.to_datetime(ob_df['주문일시'])
```

```python
fs_df['결제 완료일'] = pd.to_datetime(fs_df['결제 완료일'])
fs_df['결제 완료일'] = fs_df['결제 완료일'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
fs_df['결제 완료일'] = pd.to_datetime(fs_df['결제 완료일'])
fs_df.rename(columns={'결제 완료일': '주문일시'}, inplace=True)

fs_df = fs_df[fs_df['주문일시'] >= datetime(2025, 2, 3)]
fs_df = fs_df[fs_df['주문일시'] <= datetime(2025, 3, 19)]
```

```python
fs_df.drop(fs_df[fs_df['상품명'].str.contains('테스트')].index, inplace=True)
```

```python
status = ['결제완료', '배송중', '배송완료', '배송준비', '구매확정']
fs_df = fs_df[fs_df['주문 상태'].isin(status)]
```

```python
fs_df = fs_df.reset_index()
```

```python
datas = []

for i in fs_df.index:
    txt = fs_df.loc[i, '상품명']
    if '테스트' in txt:
        data = '옵션확인필요'
        datas.append(data)
    elif '임직원' in txt:
        data = '옵션확인필요'
        datas.append(data)
    elif '스페셜' in txt:
        data = '옵션확인필요'
        datas.append(data)
    else:
        data = txt.split('/')[0]
        data = data.split()[-1]
        datas.append(data)

fs_df['product'] = datas
```

```python
for i in fs_df[fs_df['product'] == '옵션확인필요'].index:
    fs_df.loc[i, 'product'] = fs_df.loc[i, '옵션명']
```

```python
fs_df['product'] = fs_df['product'].apply(lambda x: x[:2])
fs_df['product'].unique()
```

```python
ob_df['품목명'] = ob_df['품목명'].apply(lambda x: x[:2])
ob_df['품목명'].unique()
```

```python
fs_df.loc[fs_df['product'].str.contains('에스'), 'product'] = '프로'
```

```python
fs_df['product'].unique()
```

```python
ob_df['product'] = ob_df['품목명']
ob_df
```

```python
df_inner = pd.merge(ob_df, fs_df[['주문일시', '상품명', '옵션명', 'product']], on=["주문일시", 'product'], how="left")
```
