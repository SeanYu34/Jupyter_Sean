```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.simplefilter('ignore')

df = pd.read_excel('/Users/seanyu_mac/Downloads/주문내역_2024-03-19~2025-03-19.xlsx')
status = [50557, 51174, 51181, 51176]
df = df[df['상품 번호'].isin(status)]
status = ['결제완료', '배송중', '배송완료', '구매확정']
df = df[df['주문 상태'].isin(status)]

for i in df['상품명'].unique():
    con = df['상품명'] == i
    start_date = df.loc[con, '결제 완료일'].min()
    start_date.strftime('%Y-%m-%d')
    print(f'{i} : {start_date}')
```
