```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

df = pd.read_csv('/Users/seanyu_mac/Downloads/public_admin_group_buy_excel_order_export_2025-02-26_162550.csv')
df
```

```python
df['reg_dt'] = pd.to_datetime(df['reg_dt'])
time = df['reg_dt'] + timedelta(hours=9)
df['time'] = time
```

```python
print('주간 파마브로스 전체 공동구매 주문 건수는')
df.groupby(pd.Grouper(key='time', freq='W'))['order_no'].size()[75:]
```

```python
print('주간 파마브로스 전체 공동구매 주문 금액은')
df.groupby(pd.Grouper(key='time', freq='W'))['order_price'].sum()[75:]
```
