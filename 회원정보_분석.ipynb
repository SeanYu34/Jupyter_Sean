```python
user_df['first_seen'] = pd.to_datetime(user_df['first_seen'])
user_df['last_seen'] = pd.to_datetime(user_df['last_seen'])
user_df['signup_month'] = user_df['first_seen'].dt.to_period('M')
user_df['last_active_month'] = user_df['last_seen'].dt.to_period('M')

# Calculate retention: how many users were active in a given month after signing up
retention = (
    user_df.groupby('signup_month')
    .apply(lambda x: pd.Series({
        'signups': x['user_id'].nunique(),
        'retained_next_month': x.loc[x['last_active_month'] > x['signup_month'], 'user_id'].nunique()
    }))
)

# Calculate retention rate as a percentage
retention['retention_rate'] = (retention['retained_next_month'] / retention['signups']) * 100
retention
```
