```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/seanyu_mac/Downloads/[APP] 상담 정보 데이터_기저질환&의약품_2024.12의 사본 - Raw.csv')
df
```

```python
# for i in df['underlying_conditions (기저질환)']
df['underlying_conditions (기저질환)'].unique().tolist()
```

```python
df = df.dropna()

diseases = ['혈압', '당뇨', '고지혈', '간질환', '갑상선', '위장질환', '빈혈', 'B형간염', '과민성대장증후군','아토피','자궁근종','녹내장', '요로결석','골다공증', '간', '위장', '위','간염', '요로','혈당', '과민성', '대장', '눈', '안구', '뼈', '자궁', '다낭성']
counts = []
for i in diseases:
    con = df['underlying_conditions (기저질환)'].str.contains(i)
    count = df[con].index.size
    counts.append(count)

dict = {}
for i in range(len(diseases)):
    dict[diseases[i]] = counts[i]
dict
```

```python
result = pd.DataFrame.from_dict(dict, orient='index')
result.rename(columns={0: '기저질환 언급된 횟수'}, inplace=True)
result.sort_values(by='기저질환 언급된 횟수', ascending=False)
```
